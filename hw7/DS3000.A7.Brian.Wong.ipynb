## Question 6: 

(15 pts) Predict poultry consumption from 2015 to 2020 using a simple linear regression model.
1) Ensure that you visualize the original dataset along with the model's predictions.
2) Evaluate the model's predictions against the actual consumption data from 2015 to 2020 and discuss its performance.

Question 6.1

```python
import numpy as np
from sklearn.linear_model import LinearRegression

X = df_filtered["TIME"].to_numpy().reshape(-1, 1)  # Explanatory variable (Years)
y = df_filtered["Value"].to_numpy().reshape(-1, 1)  # Response variable (Consumption)

lr_model = LinearRegression(fit_intercept=True) # initialize the regression model
lr_model = lr_model.fit(X, y)                   # find the line of best fit

#display the parameters for the slope and intercept
print('The intercept is: ', lr_model.intercept_[0])
print('The slope is: ', lr_model.coef_[0,0])

future_years = np.array([2015, 2016, 2017, 2018, 2019, 2020]).reshape(-1, 1)
predicted_values = lr_model.predict(future_years)

plt.figure(figsize=(10, 5))

# Regression line (1994-2014 training data)
sns.regplot(x=df_filtered["TIME"], y=df_filtered["Value"], label="Trend (1994-2014)")

# Scatter plot for actual values (2015-2020)
df_actual = df[(df["LOCATION"] == "MEX") & 
               (df["SUBJECT"] == "POULTRY") & 
               (df["MEASURE"] == "THND_TONNE") & 
               (df["TIME"] >= 2015) & 
               (df["TIME"] <= 2020)][["TIME", "Value"]]

sns.scatterplot(x=df_actual["TIME"], y=df_actual["Value"], color='green', label="Actual (2015-2020)", marker="o", s=100)

# Scatter plot for predicted values (2015-2020)
future_years = np.array([2015, 2016, 2017, 2018, 2019, 2020])
sns.scatterplot(x=future_years, y=predicted_values.flatten(), color='red', label="Predicted (2015-2020)", marker="x", s=100)

# Formatting the plot
plt.title("Poultry Consumption in Mexico: Predictions (2015-2020)")
plt.xlabel("Year")
plt.ylabel("Consumption (Thousand Tonnes)")
plt.legend()
plt.grid(True)
plt.show()
```

Question 6.2

The model's predictions against the actual consumption data from 2015 to 2020 is consistently about only 100 consumption (thousand tonnes) above the actual values, which is a small difference. Considering how close the predicted values are to the actual consumption data from 2015 to 2020, I would say that the model's prediction was a good performance as it corrected showed the trend of consumption from 2015 to 2020--an upward trend that is increasing at a near linear rate--and the predicted values are only a small margin away from the actual values.

## Question 7: 

(5 pts) Calculate the R-squared and explain what this tells you about the model.

```python
#get r-squared, i.e. the coefficient of determination
r_sqrd = lr_model.score(X, y)
r_sqrd = round(r_sqrd, 2)

print('r-squared is: ', r_sqrd)
```

The r-squared value is 0.98, which is very close to 1. This tells me that the model is very good at predicting the data and is a near perfect fit for the data as 0.98 is only 0.02 off from 1--a perfect fit.

## Question 8: 

(5 pts)[optional/bonus] This is a challenge question that may require some research.
<b>residuals</b>: Visualize the residuals of the regression model in question 6 and explain the chart.
Note: A residual is the difference between the actual and predicted values.

```python
y_pred = lr_model.predict(X)

# Calculate residuals
residuals = y - y_pred

# Visualize residuals
plt.figure(figsize=(10, 5))
sns.scatterplot(x=df_filtered["TIME"], y=residuals.flatten(), color='blue', label="Residuals")
plt.title("Residuals of the Regression Model")
plt.xlabel("Year")
plt.ylabel("Residuals (Actual - Predicted)")
plt.grid(True)
plt.legend()
plt.show()
```

In the chart "Residuals of the Regression Model," the blue dots represent the residuals, the x-axis represents the year, and the y-axis represents the residuals (actual - predicted). In this residuals graph, the absolute difference for a residual is just over 200, which is not bad as our measurement for consumption is measured in thousands, so an offset of 200 is still very close. A trend I see is that at first at around 1995, the residuals start close to -200 to -50, then as the year increases, the residuals also increases almost linearly to a peak of just over 200 at 2005. After 2005, the residuals start to decrease as the year increases at an almost linearly rate back to about -150 by 2014.
