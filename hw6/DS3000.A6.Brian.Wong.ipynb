## Question 5: 

(10 pts) Calculate the Pearson Correlation coefficient of <b>carat</b> and <b>price</b>. Explain the results and discuss the strength of the correlation.

```python
from scipy import stats

# Compute Pearson correlation coefficient and p-value
corr, p_value = stats.pearsonr(df["carat"], df["price"])

# Print values
print('The correlation coefficient is: ', corr)
print('The p-value is: ', p_value)
```

THe correlation coefficient is 0.922, indicating a strong positive correlation between carat and price, so as carat increases, price also tends to increase. The p-value is 0, indicating that its unlikely that this pattern would appear if no correlation was evident.
The strength of the correlation is a very strong positive correlation because the correlation coefficient is 0.922, which is very close to 1.

## Question 7: 

(10 pts) Using a 2-sample t-test**, determine if there is a statistical difference between the <b>price</b> of diamonds with a <b>cut</b> that is considered:
1) 'Good' vs 'Very Good'
2) 'Premium' vs 'Ideal'
<br> <b>**set alpha to 5% i.e. 0.05.</b></br>

```python
# Define the two groups for each comparison
good_cut_prices = df[df["cut"] == "Good"]["price"]
very_good_cut_prices = df[df["cut"] == "Very Good"]["price"]

premium_cut_prices = df[df["cut"] == "Premium"]["price"]
ideal_cut_prices = df[df["cut"] == "Ideal"]["price"]

alpha = 0.05
# Perform the two-sample t-test for 'Good' vs 'Very Good'
t_stat_good_vs_very_good, p_val_good_vs_very_good = stats.ttest_ind(good_cut_prices, very_good_cut_prices)

# Perform the two-sample t-test for 'Premium' vs 'Ideal'
t_stat_premium_vs_ideal, p_val_premium_vs_ideal = stats.ttest_ind(premium_cut_prices, ideal_cut_prices)

#compare the p_value with alpha to determine the significance
if(p_val_good_vs_very_good < alpha):
    print('The price between "Good" cuts and "Very Good" cuts are statistically different.')
else:
    print('The price between "Good" cuts and "Very Good" cuts are not statistically different')

if(p_val_premium_vs_ideal < alpha):
    print('The price between "Premium" cuts and "Ideal" cuts are statistically different.')
else:
    print('The price between "Premium" cuts and "Ideal" cuts are not statistically different')
```
